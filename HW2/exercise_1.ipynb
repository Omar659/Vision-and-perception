###Let $l$ be the line of interest;

I found the three points that lie on $l$ using the cross product between $l$ and the other three lines: $l_1$, $l_2$ and $l_3$, obtaining $x_1$, $x_2$ and $x_3$.

Each point and line was normalised by dividing it by the third value (the *z-value*, in order to have it at **1**).

```python
# line l
l = np.array([1,2,3])

# three lines l1, l2 and l3 to obtain the points
l1 = np.array([3,2,5])
l2 = np.array([3,3,3])
l3 = np.array([5,7,2])

# three collinear points x1, x2 and x3 
x1 = np.cross(l, l1)
x2 = np.cross(l, l2)
x3 = np.cross(l, l3)

# normalization dividing by third value (z-value) in order to obtain 1 as third value
l = l/l[2]
l1 = l1/l1[2]
l2 = l2/l2[2]
l3 = l3/l3[2]
x1 = x1/x1[2]
x2 = x2/x2[2]
x3 = x3/x3[2]

# vector of x coordinate of the three collinear points
x = [x1[0], x2[0], x3[0]]
# vector of y coordinate of the three collinear points
y = [x1[1], x2[1], x3[1]]
```

###Plot of lines and points on a *3D* and *2D* plot:
1.   The *3D* plot shows the lines in $‚Ñù^3$ starting from the **origin** and intersecting the **z = 1 plane**. The lines corresponding to points $x_i$ are the result of the cross product between $l$ and $l_i$. The result of the product is shown on the **z = 1 plane** with the three points $x_i$. Then a line is drawn between the points to show that they are collinear.
2.   The *2D* plot, instead, shows a top view of the **z = 1 plane** to highlight the three points found to be collinear.

```python
# N.B. All lines have been extended by multiplying the plotted values by 1.2 in order to better see the intersections.
fig = plt.figure(figsize=(6.4*3, 4.8*3))

# First subplot: 2D
ax = fig.add_subplot(1, 2, 1, title = "2d representation at z=1")

ax.plot(x, y, color = 'black', label = "Line of three point")
ax.scatter(x,y, color = "b", label = "Points x1, x2, x3")
ax.scatter(l[0],l[1], color = [1,0,1])
ax.scatter(l1[0],l1[1], color = [1,0,1])
ax.scatter(l2[0],l2[1], color = [1,0,1])
ax.scatter(l3[0],l3[1], color = [1,0,1])
ax.plot([0, l[0]*1.2], [0, l[1]*1.2], color = 'y', label = "Line l")
ax.plot([0, l1[0]*1.2], [0, l1[1]*1.2], color = [1,0,0], label = "Line l1")
ax.plot([0, l2[0]*1.2], [0, l2[1]*1.2], color = [0.8,0,0], label = "Line l2")
ax.plot([0, l3[0]*1.2], [0, l3[1]*1.2], color = [0.6,0,0], label = "Line l3")
ax.plot([0, x1[0]*1.2], [0, x1[1]*1.2], color = [0,1,0], label = "Line l x l1")
ax.plot([0, x2[0]*1.2], [0, x2[1]*1.2], color = [0,0.8,0], label = "Line l x l2")
ax.plot([0, x3[0]*1.2], [0, x3[1]*1.2], color = [0,0.6,0], label = "Line l x l3")

ax.grid(True)
plt.legend()

# Second subplot: 3D
ax = fig.add_subplot(1, 2, 2, projection='3d', title = "3d representation")

# The plane z=1
# The values for the size of the plane to be shown were chosen by looking at the 2D plot for a clean representation
# The plane z=1 in reality includes all possible points
lX = np.linspace(-1.5,7, 10) #the value is chosen
lY = np.linspace(-5,5, 10)
pX,pY = np.meshgrid(lX,lY)
pZ = np.ones((1,10))

ax.plot(x, y, zs = 1, color = 'black', label = "Line of three point")
ax.plot([0, l[0]*1.2], [0, l[1]*1.2], [0, l[2]*1.2], color = 'y', label = "Line l")
ax.plot([0, l1[0]*1.2], [0, l1[1]*1.2], [0, l1[2]*1.2], color = [1,0,0], label = "Line l1")
ax.plot([0, l2[0]*1.2], [0, l2[1]*1.2], [0, l2[2]*1.2], color = [0.8,0,0], label = "Line l2")
ax.plot([0, l3[0]*1.2], [0, l3[1]*1.2], [0, l3[2]*1.2], color = [0.6,0,0], label = "Line l3")
ax.plot([0, x1[0]*1.2], [0, x1[1]*1.2], [0, x1[2]*1.2], color = [0,1,0], label = "Line l x l1")
ax.plot([0, x2[0]*1.2], [0, x2[1]*1.2], [0, x2[2]*1.2], color = [0,0.8,0], label = "Line l x l2")
ax.plot([0, x3[0]*1.2], [0, x3[1]*1.2], [0, x3[2]*1.2], color = [0,0.6,0], label = "Line l x l3")
ax.scatter(x,y, [1,1,1], color = "b", label = "Points x1, x2, x3")
ax.scatter(l[0],l[1], [1], color = [1,0,1])
ax.scatter(l1[0],l1[1], [1], color = [1,0,1])
ax.scatter(l2[0],l2[1], [1], color = [1,0,1])
ax.scatter(l3[0],l3[1], [1], color = [1,0,1])
ax.plot_surface(pX, pY, pZ, color = 'b', alpha = 0.1, label = "Plane z=1")

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)

plt.show()
# print(x,y)
```

###Demonstration
To show that the points obtained lie on $l$ we need to do the dot product between $l^T$ and the point $x_i$, if this is **0** then it is a point on the line, otherwise not.

In the print we can see that the values obtained are: **0** or **very small values tending towards 0** - *most likely due to some rounding up done* - and therefore it is a result that meets the requirement.

```python
# dot product between l transpose and x1, x2 and x3
prove1 = np.dot(l.T, x1)
prove2 = np.dot(l.T, x2)
prove3 = np.dot(l.T, x3)

# proves printed
if (prove1 == 0):
  print("l.T*x1 = " + str(prove1))
else:
  print("l.T*x1 = " + str(prove1) + " ‚âÖ 0")

if (prove2 == 0):
  print("l.T*x2 = " + str(prove2))
else:
  print("l.T*x2 = " + str(prove2) + " ‚âÖ 0")
  
if (prove3 == 0):
  print("l.T*x3 = " + str(prove3))
else:
  print("l.T*x3 = " + str(prove3) + " ‚âÖ 0")
```

###Projective point transformation

I used the non-singular matrix **$H$** to map the points **$x_1$**, **$x_2$** and **$x_3$** to other points **$x'_1$**, **$x'_2$** and **$x'_3$** and the line **$l$** to another line **$l'$**.

To map the line $l$ into $l'$ I used the formula $l' = H^{-T}l$.

To map the points into a single calculation I created the matrix $X = [x_1^T\ x_2^T\ x_3^T]$ and the matrix $X' = HX$. The columns of $X'$ are the three new points mapped to $H$.
Then, all the new points and the new line were **normalized** by dividing by their *third value*.

Finally, the *x-coordinates* and *y-coordinates* of the points were inserted into two new vectors for plotting.



```python
# Matrix H
H = [[0.02,0.35,0.02], 
     [0.05,0.16,0.22], 
     [0.41,0.48,0.19]]

print("H must be non-singular: determinant(H) = " + str(np.linalg.det(np.array(H))) + " ‚â† 0")
```

```python
# Matrix X with x1, x2 and x3 as column
X = np.array([[x1[0], x2[0], x3[0]],
              [x1[1], x2[1], x3[1]],
              [x1[2], x2[2], x3[2]]])

# l' and X' mapped with H. 
#I used l.T because for a correct calculation I needed the line l to be already transposed.
lp = np.dot(np.linalg.inv(H).T, l.T)
Xp = np.dot(H, X)

# x'1, x'2 and x'3
x1p = Xp[:, 0]
x2p = Xp[:, 1]
x3p = Xp[:, 2]

# Normalization dividing by third value
x1p = x1p/x1p[2]
x2p = x2p/x2p[2]
x3p = x3p/x3p[2]
lp = lp/lp[2]

# vector of x coordinate of the new three collinear points mapped with H
xp = [x1p[0], x2p[0], x3p[0]]
# vector of y coordinate of the new three collinear points mapped with H
yp = [x1p[1], x2p[1], x3p[1]]
```

###Plot of $l$ and $l'$ and points $x_1$, $x_2$, $x_3$, $x'_1$, $x'_2$ and $x'_3$ on a *2D* plot
The *2D* plot shows a top view of the **z = 1 plane** of the points and lines mentioned in the title. 

It can be seen that the collinearity and proportional distances of the three points are preserved even after applying the H

```python
fig = plt.figure(figsize=(6.4*2, 4.8*2))

ax = fig.add_subplot(title = "2d representation at z=1")

ax.plot(x, y, color = [0,1,0], label = "Line of three point")
ax.plot(xp, yp, color = [0,0,1], label = "Line of three point mapped with H")
ax.scatter(x, y, color = 'g', label = "Points x1, x2, x3")
ax.scatter(xp, yp, color = "b", label = "Points x'1, x'2, x'3")
 
ax.grid(True)
plt.legend()
plt.show()

# This print is used to show that the line l in this graph and the previous one 
# are the same and it is the different size of the plot that makes them look different. 
# print(x,y)
```

###Demonstration
To show that the new points $x'_1$, $x'_2$ and , $x'_3$ lie on the line $l'$ all calculated with $H$ we need to do the dot product between $l'^T$ and the point $x'_i$, if this is **0** then it is a point on the line, otherwise not.

In the print we can see that the values obtained are: **0** or **very small values tending towards 0** - *most likely due to some rounding up done* - and therefore it is a result that meets the requirement.

```python
# dot product between l' transpose and x'1, x'2 and x'3
prove1p = np.dot(lp.T, x1p)
prove2p = np.dot(lp.T, x2p)
prove3p = np.dot(lp.T, x3p)

# proves printed
if (prove1p == 0):
  print("l'.T*x'1 = " + str(prove1p))
else:
  print("l'.T*x'1 = " + str(prove1p) + " ‚âÖ 0")

if (prove2p == 0):
  print("l'.T*x'2 = " + str(prove2p))
else:
  print("l'.T*x'2 = " + str(prove2p) + " ‚âÖ 0")
  
if (prove3p == 0):
  print("l'.T*x'3 = " + str(prove3p))
else:
  print("l'.T*x'3 = " + str(prove3p) + " ‚âÖ 0")
```

#Exercise 1.1 with symbols
Demonstration that any choice of lines $l$, $l_1$, $l_2$ and $l_3$ initially would not have affected the calculation

```python
# symbols used
(l_1, l_2,l_3,
 l1_1, l1_2, l1_3,
 l2_1, l2_2, l2_3,
 l3_1, l3_2, l3_3) = symbols("l_1,l_2,l_3," + 
                           "l1_1,l1_2,l1_3," +
                           "l2_1,l2_2,l2_3," +
                           "l3_1,l3_2,l3_3", real =True)
 
line = np.array([l_1, l_2, l_3])
line1 = np.array([l1_1, l1_2, l1_3])
line2 = np.array([l2_1, l2_2, l2_3])
line3 = np.array([l3_1, l3_2, l3_3])
print("The line l:\nl = " + str(line) + "\n")
print("The line l1, l2 and l3:")
print("l1 = " + str(line1))
print("l2 = " + str(line2))
print("l3 = " + str(line3))

line = line/line[2]
line1 = line1/line1[2]
line2 = line2/line2[2]
line3 = line3/line3[2]
line = simplify(line)
line1 = simplify(line1)
line2 = simplify(line2)
line3 = simplify(line3)

print("\nThe line l1, l2 and l3 after normalization:")
print("l = l/l[2] = " + str(line))
print("l1 = l1/l1[2] = " + str(line1))
print("l2 = l2/l2[2] = " + str(line2))
print("l3 = l3/l3[2] = " + str(line3))

point_x1 = np.cross(line,line1)
point_x2 = np.cross(line,line2)
point_x3 = np.cross(line,line3)
point_x1 = simplify(point_x1)
point_x2 = simplify(point_x2)
point_x3 = simplify(point_x3)
print("\nThe point x1, x2 and x3:")
print("x1 = l x l1 = " + str(point_x1))
print("x2 = l x l2 = " + str(point_x2))
print("x3 = l x l3 = " + str(point_x3))

point_x1 = point_x1/point_x1[2]
point_x2 = point_x2/point_x2[2]
point_x3 = point_x3/point_x3[2]
point_x1 = simplify(point_x1)
point_x2 = simplify(point_x2)
point_x3 = simplify(point_x3)
print("\nThe point x1, x2 and x3 after normalization:")
print("x1 = x1/x1[2] = " + str(point_x1))
print("x2 = x2/x2[2] = " + str(point_x2))
print("x3 = x3/x3[2] = " + str(point_x3))

sprove1 = np.dot(line, point_x1)
sprove2 = np.dot(line, point_x2)
sprove3 = np.dot(line, point_x3)
print("\nProve that x1 lie on l:")
print("l.T * x1 = " + str(sprove1) + "\nWith semplification l.T * x1 = " + str(simplify(sprove1)))
print("\nProve that x2 lie on l:")
print("l.T * x2 = " + str(sprove2) + "\nWith semplification l.T * x2 = " + str(simplify(sprove2)))
print("\nProve that x3 lie on l:")
print("l.T * x3 = " + str(sprove3) + "\nWith semplification l.T * x3 = " + str(simplify(sprove3)))
```
