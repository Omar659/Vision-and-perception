###Symbols used

```python
# Symbols used
(u1,u2,
 v1,v2,
 a11,a12,a13,
 a21,a22,a23,
 x1,x2,x3,
 m) = symbols("u1,u2,v1,v2,a11,a12,a13,a21,a22,a23,x1,x2,x3,m", real =True) 
```

###Affine transformation


```python
Ha = Matrix(3,3,[a11, a12, a13,
                 a21, a22, a23,
                 0, 0, 1])
Ha
```

###Generic transformation with $X \to X';\ X' = H_aX$

Point X

```python
X = np.array([x1,x2,x3])
Array(X)
```

Point X'

```python
Xp = Ha@X
Array(Xp)
```

###We now define two points $U$ and $V$ for our segment

```python
U = np.array([u1,u2,1])
V = np.array([v1,v2,1])
print(Array(U), Array(V))
```

###And the two points $U'$ and $V'$ with transformation $H_a$

Point $U'$

```python
Up = Array(Xp).subs(x1, U[0]).subs(x2,U[1]).subs(x3,U[2])
Up
```

Point $V'$

```python
Vp = Array(Xp).subs(x1, V[0]).subs(x2,V[1]).subs(x3,V[2])
Vp
```

###$V'-U'$ to obtain the segment $U'V'$

```python
UpVp = Vp - Up
UpVp
```

###Now the square distance of $U'V'$

```python
UpVp2 = np.dot(UpVp, UpVp)
UpVp2
```

A few arithmetical steps to obtain the final result

```python
UpVp2 = collect(UpVp2, [a11,a12,a21,a22])
UpVp2
```

```python
UpVp2 = expand(UpVp2, mul = false)
UpVp2
```

```python
UpVp2 = UpVp2.subs((u1**2 - 2*u1*v1 + v1**2), (v1-u1)**2).subs((u2**2 - 2*u2*v2 + v2**2), (v2-u2)**2)
UpVp2
```

```python
UpVp2 = collect(UpVp2, [(-u1+v1)*(-u2+v2), (v1-u1), (v2-u2)])
UpVp2
```

```python
UpVp2 = rcollect(UpVp2, 2)
UpVp2
```

###Now for the original points: $V-U$ to obtain the segment $UV$

```python
UV = V - U
Array(UV)
```

###Square distance of $UV$

```python
UV2 = np.dot(UV, UV)
UV2
```

###Ratio

The following ratio should be the same for parallel line segments:
 
$\frac{(U'V')^2}{(UV)^2}$

In fact if we had two other points $S$ and $T$ and the relative segment $ST$ (and therefore also $S'$ and $T'$ and $S'T'$) we would obtain the proportion:

$\frac{(U'V')^2}{(UV)^2} = \frac{(S'T')^2}{(ST)^2}$

And by the properties of proportions we can invert the middle, finally obtaining that:

$\frac{(U'V')^2}{(S'T')^2} = \frac{(UV)^2}{(ST)^2}$

It is the ratio of the length of the two segment lines before and after the affine transformation.

So if we find a poperty such that with the segment $ST$ (and $S'T'$), being parallel to $UV$ (and $U'V'$), we get $\frac{(U'V')^2}{(UV)^2}$ then it is demonstrated that the ratio is preserved

###$R = \frac{(U'V')^2}{(UV)^2}$

```python
R = UpVp2/UV2
R
```

If $u_1 = v_1$ then the segment is vertical and the formula written above is simplified and only $(a_{12}^2+a_{22}^2)$ remains. Being terms of the affine transformation, any other vertical (parallel) line you use will give the same result and so the ratio will be preserved.

If $u_1 â‰  v_1$ we can think of the angular coefficient of the line segment as $m = \frac{v_2-u_2}{v_1-u_1}$

First we divide each term by $(v_1-u_1)^2$ and we get:

```python
Rm = (expand(UpVp2/((v1-u1)**2), multinomial=false, deep = false))/(expand(UV2/((v1-u1)**2), multinomial=false, deep = false))
Rm
```

Then we substitute $m$ for the various $\frac{v_2-u_2}{v_1-u_1}$ terms that appear and we obtain:

```python
Rm = Rm.subs((v2 - u2)/(v1 - u1), m)
Rm
```

The above formula depends on terms in the affine transformation and the angular coefficient of the line segment. This means that the formula varies depending on the value of m. However, if we take another parallel line segment, this one, being parallel, will have the same angular coefficient $m$ and therefore obtain the same ratio, preserving it

```python
# Symbols used
(x, y, z, a, b, c, d, e, f) = symbols("x, y, z, a, b, c, d, e, f", real =True) 
```

We start with the equation of the **conic**

```python
# Conic equation
conic = a*x**2 + b*x*y + c*y**2 + d*x*z + e*y*z + f*z**2
conicEq = Eq(conic,0)
conicEq
```

If we choose $b = 0$ and $a = c$ and finally $c = 1$ we obtain the equation of a circle with **three unknowns**

```python
# Circle equation given by the conic equation
circleByConicEq = conicEq.subs(a, c).subs(b, 0).subs(c,1)
circleByConicEq
```

To obtain the **unknowns** of the above formula, having **three** of them, you need **three** equations given by **three** points. 

But we also know that given any **circumference equation**, it will intersect, symmetrically, the **line at infinity** $l_\infty$ at the two circular points $I=(1,\ +i,\ 0)^T, J=(1,\ -i,\ 0)^T$.

```python
# Circular points
I = np.array([1, +1j, 0])
J = np.array([1, -1j, 0])

# Substitutions
print("substitution with circular point I:\n" + str(circleByConicEq.lhs.subs(x, I[0]).subs(y, I[1]).subs(z, I[2])) + " = 0")
print("substitution with circular point J:\n" + str(circleByConicEq.lhs.subs(x, J[0]).subs(y, J[1]).subs(z, J[2])) + " = 0")
```

All this is to say that given the equation of the **conic** transformed into the equation of a **circle**, **three points** are sufficient to define a circle, but each circle intersects the line at infinity in its **two circular points**, and so:

**"Starting from a conic, having three points and two circular points we define a circle."**

---
**Numerical example:**

We define the **system of five equations** given by the conic formula by replacing the coordinates with those of the **three points** and the **two circular points**

```python
# Three points plus I and J defined above
point1 = [17, 13, 15]
point2 = [56, 44, 25]
point3 = [2, 64, 70]

# Five equations constructed from the conic equation defined above by substitution with coordinates of points
eq1 = conic.subs(x, I[0]).subs(y, I[1]).subs(z, I[2])
eq2 = conic.subs(x, J[0]).subs(y, J[1]).subs(z, J[2])
eq3 = conic.subs(x, point1[0]).subs(y, point1[1]).subs(z, point1[2])
eq4 = conic.subs(x, point2[0]).subs(y, point2[1]).subs(z, point2[2])
eq5 = conic.subs(x, point3[0]).subs(y, point3[1]).subs(z, point3[2])

# System of the five equations
conicSystem = Matrix([eq1, eq2, eq3, eq4, eq5])
conicSystem
```

From the solution of the system we can see that we have $b = 0$ and $a = c$ as mentioned above. Having $b = 0$ and $a = c$ we will have that from an equation of the conic we obtain that of a circle.

However, each variable is multiplied by $f$

```python
# Solution of the system
solve(conicSystem)
```

To avoid each unknown being multiplied by $f$, replace it in the system with $f = 1$

```python
# New system with f = 1
conicSystem = conicSystem.subs(f, 1)
conicSystem
```

Finally, we have a solution without $f$ and which respects the constraint $b = 0$ and $a = c$ of the formula for the circle given a conic

```python
# Solution of this new system
solve(conicSystem)
```
